### Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.
### Answer : Probability Mass Function (PMF) and Probability Density Function (PDF) are mathematical concepts used in probability theory and statistics to describe the likelihood of different outcomes in a random variable.

### 1. Probability Mass Function (PMF):

### PMF is applicable to discrete random variables, which take on distinct, separate values.
### It gives the probability of each possible outcome.
### Mathematically, for a discrete random variable X, the PMF is denoted by P(X = x), where x is a specific value that X can take.
### The sum of all probabilities in the PMF equals 1.
### Example:
### Consider a fair six-sided die. The PMF for the outcome of rolling the die can be expressed as:

### P(X = 1) = 1/6
### P(X = 2) = 1/6
### P(X = 3) = 1/6
### P(X = 4) = 1/6
### P(X = 5) = 1/6
### P(X = 6) = 1/6
### 2. Probability Density Function (PDF):

### PDF is applicable to continuous random variables, which can take any value within a range.
### It represents the relative likelihood that a continuous random variable will take on a particular value.
### Unlike PMF, where you sum probabilities, the integral of the PDF over its entire range equals 1.
### Example:
### Consider a uniform distribution between 0 and 1. The PDF for this distribution is constant within this range, and outside this range, it is zero.

### For 0 <= x <= 1, the PDF f(x) is a constant, say 1.
### Outside this range, f(x) = 0.
### The integral of f(x) from -∞ to ∞ is 1.
### In summary, PMF is used for discrete random variables, providing probabilities for specific outcomes, while PDF is used for continuous random variables, giving the likelihood of the variable falling within a particular range.

### Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?
### Answer : Cumulative Density Function (CDF):

### The Cumulative Density Function (CDF) is a function associated with a probability distribution. For a random variable X, the CDF at a specific value x is the probability that X takes a value less than or equal to x. Mathematically, it is often denoted as F(x) or P(X ≤ x).

### For a discrete random variable, the CDF is obtained by summing up the probabilities from the PMF up to a particular point. For a continuous random variable, it is obtained by integrating the PDF up to that point.

### Example:

### Consider a fair six-sided die. The PMF for the die is as follows:

### P(X = 1) = 1/6
### P(X = 2) = 1/6
### P(X = 3) = 1/6
### P(X = 4) = 1/6
### P(X = 5) = 1/6
### P(X = 6) = 1/6
### The CDF for this die would be:

### F(1) = P(X ≤ 1) = 1/6
### F(2) = P(X ≤ 2) = 2/6 = 1/3
### F(3) = P(X ≤ 3) = 3/6 = 1/2
### F(4) = P(X ≤ 4) = 4/6 = 2/3
### F(5) = P(X ≤ 5) = 5/6
### F(6) = P(X ≤ 6) = 6/6 = 1
### Why CDF is used?

### 1. Probability Calculation: CDF is useful for calculating probabilities associated with a random variable. Given a value x, the CDF provides the probability that the random variable is less than or equal to x.

### 2. Understanding Distribution Characteristics: CDF provides insights into the overall distribution of the random variable. It shows how the probability accumulates as the variable takes on different values.

### 3. Quantile Calculation: CDF is essential for calculating quantiles, which represent the values at which a specified cumulative probability is reached.

### 4. Comparison of Distributions: CDFs can be used to compare different probability distributions, helping in the comparison of random variables and their associated likelihoods.

### In summary, the Cumulative Density Function is a valuable tool in probability theory and statistics, providing a way to understand and analyze the distribution of random variables and calculate probabilities associated with them.

### 6. Body Temperature: Human body temperature is approximately normally distributed, with the majority of people having temperatures around the mean.

### Parameters of the Normal Distribution:

### The normal distribution is characterized by two parameters: the mean (μ) and the standard deviation (σ). These parameters determine the shape and characteristics of the distribution.

### 1. Mean (μ):

### The mean represents the central location or the average of the distribution.
### It determines the position of the peak of the curve.
### Shifting the mean to the right or left changes the location of the center of the distribution.
### 2. Standard Deviation (σ):

### The standard deviation measures the spread or variability of the distribution.
### A larger standard deviation leads to a wider and flatter distribution, indicating greater variability.
### A smaller standard deviation results in a narrower and taller distribution, indicating less variability.
### Relationship between Parameters and Distribution Shape:

### If μ increases or decreases:

### The entire distribution shifts to the right or left along the x-axis, affecting the location of the peak.
### If σ increases:

### The distribution becomes wider and flatter, indicating greater variability.
### If σ decreases:

### The distribution becomes narrower and taller, indicating less variability.
### In summary, the mean and standard deviation are crucial parameters that define the normal distribution. They control the central tendency and spread of the data, respectively, and understanding their roles helps interpret the shape and characteristics of the distribution in various real-world scenarios.
